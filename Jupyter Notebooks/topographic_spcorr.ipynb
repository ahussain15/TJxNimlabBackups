```python
import nibabel as ni
import numpy as np
import scipy.stats as ss
import seaborn as sb
```

```python
labels = ["hospital", "acute", "3mo", "year", "mgh"]
```

### VLSM

```python
hospital_vlsm = ni.load("vlsm.nii")
acute_vlsm = ni.load("vlsm_acute.nii")
thmo_vlsm = ni.load("vlsm_thmo.nii")
year_vlsm = ni.load("vlsm_year.nii")
mgh_vlsm = ni.load("vlsm_mgh.nii")
all_vlsm = [hospital_vlsm, acute_vlsm, thmo_vlsm, year_vlsm, mgh_vlsm]
```

```python
vlsm_data = []
for i in range(len(all_vlsm)):
    vlsm_data.append([])
    img = np.array(all_vlsm[i].dataobj)
    for x in range(len(img)):
        for y in range(len(img[x])):
            for z in range(len(img[x, y])):
                if img[x, y, z] != np.nan:
                    vlsm_data[i].append(img[x, y, z])
```

```python
vlsm_corr = np.zeros((len(all_vlsm), len(all_vlsm)))
for i in range(len(all_vlsm)):
    for ii in range(len(all_vlsm)):
        vlsm_corr[i, ii] = ss.pearsonr(vlsm_data[i], vlsm_data[ii])[0]
print(vlsm_corr)
```

```python
sb.heatmap(vlsm_corr, xticklabels = labels, yticklabels = labels, vmin=0, vmax=1, cmap="vlag", annot=True)
```
